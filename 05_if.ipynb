## <font color = "blue">bool型</font>

## <font color = "blue">比較演算子</font>
**比較演算子**：ふたつの値を比較して、True(真)またはFalse(偽)のどちらかのbool型の値を返す演算子


| 演算子  |  例    | 説明 |
| :---:   | :----: |    :----:   |
|  ==   |  a == b    | aとbの値が等しいならTrue、そうでないならFalse    |
|  !=   |  a != b    | aとbの値が等しくないならTrue、そうでないならFalse      |
|  >    |  a > b     | aがbより大きいときTrue、そうでないならFalse     |
|  >=   |  a >= b    | aがb以上のときTrue、そうでないならFalse      |
|  <    |  a < b     | aがbより小さいときTrue、そうでないならFalse     |
|  <=   |  a <= b    | aがb以下であればTrue、そうでないならFalse    |
|  in   |  a in b    | aがbの要素であればTrue、そうでないならFalse     |
| not in|  a not in b| aがbの要素でなければTrue、そうでないならFalse     |
|  is   |  a is b    | aとbが同じオブジェクトならTrue、そうでないならFalse    |
| is not|  a is not b| aとbが同じオブジェクトでないならTrue、同じオブジェクトならFalse     |



<font color = "green">**注意**</font>：「==」と「＝」の相違
「==」は左辺と右辺の値が等しいかどうかを判断
⇒「==」はbool型で戻ってくる
「=」は変数への代入

## <font color = "blue">else/elif</font>

**else**：条件式が成立しなかったときの処理を書く  
**elif**：処理を3つ以上に分けたいとき  
　　「else if」の略


<div>

</div>

```python
# フルーツをまとめて買ってくれた人には割引してあげたい！
# 今回は5個以上で2割引、10個以上で3割引してあげよう！
fruits = ["apple", "grape"]
prices = {"apple":100, "grape":50}

num_apple = 5
num_grape = 6

total_price = prices["apple"] * int(num_apple) + prices["grape"] * int(num_grape)
if int(num_apple) + int(num_grape) >= 5:
    total_price *= 0.8 # 2割引き = 0.8倍
    print("2割引しました")
elif int(num_apple) + int(num_grape) >= 10:
  total_price *= 0.7 # 3割引 = 0.7倍
  print("3割引しました")
else:
  print("割引はありません")

print("合計金額は" + str(int(total_price)) + "円です")

```

<font color = "red">**上記のものは間違い！！**</font>

合計個数が10のとき、10も5以上の値だから最初の2割引の条件でTrueになってしまっている！

if文は上から照合して<u>最初にTrueだった条件の中身が実行されて残りは実行されない！</u>

```python
num_apple = 5
num_grape = 6

total_price = prices["apple"] * int(num_apple) + prices["grape"] * int(num_grape)
if int(num_apple) + int(num_grape) >= 10:
    total_price *= 0.7 # 3割引き = 0.7倍
    print("3割引しました")
elif int(num_apple) + int(num_grape) >= 5:
  total_price *= 0.8 # 2割引 = 0.8倍
  print("2割引しました")
else:
  print("割引はありません")

print("合計金額は" + str(int(total_price)) + "円です")

```
