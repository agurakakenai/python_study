
# <font color = "blue">while文
**while文**：回数が条件によって変わる繰り返し処理  
　　　　　　　条件がTrueの限り繰り返される





input関数のおかげでプログラムを実行してから個数を入力できるように...  
だけど上記のままだと結局1回1回、実行ボタンを押さないといけない  
　　　　　　　　　　　↓  
今回はプログラムを<font color = "red">繰り返し実行できるようにして、自動で終了</font>できるようにしたい  
このように条件に基づいて繰り返しをしたい場合は<font color = "red">while文</font>が便利！  



````python
# 前のセッションでやったプログラム
fruits = ["apple", "grape"]
prices = {"apple":100, "grape":50}

try:
  num_apple = input("appleの個数を入力してください：")
  num_grape = input("grapeの個数を入力してください：")

  total_price = prices["apple"] * int(num_apple) + prices["grape"] * int(num_grape)
  if int(num_apple) + int(num_grape) >= 10:
      total_price *= 0.7 # 3割引き = 0.7倍
      print("3割引しました")
  elif int(num_apple) + int(num_grape) >= 5:
    total_price *= 0.8 # 2割引 = 0.8倍
    print("2割引しました")
  else:
    print("割引はありません")

  print("合計金額は" + str(int(total_price)) + "円です")

except ValueError:
  print("半角整数値を入力してください")
````


## <font color = "blue">while文の書き方：</font>
```
while 条件式:    #セミコロンを忘れずに！
  処理１         #インデントに注意！！
  処理２…
```


条件が<font color = "red">真(True)</font>の場合に実行され、その後再び条件を判断する  
条件が<font color = "red">偽(False)</font>の場合は、ループを終了して、その次のコードを実行  

<font color = "green">**注意**</font>：もしwhileの条件が常に真(True)のとき、無限ループが発生  
　　　（ループが永遠に繰り返されてしまう）






````python
# runningが1の間、繰り返されるwhile文を作る
fruits = ["apple", "grape"]
prices = {"apple":100, "grape":50}

running = 1 # whileの条件用の変数を定義（boolで定義してもok）
while running == 1:
  try:
    num_apple = input("appleの個数を入力してください：")
    num_grape = input("grapeの個数を入力してください：")

    total_price = prices["apple"] * int(num_apple) + prices["grape"] * int(num_grape)
    if int(num_apple) + int(num_grape) >= 10:
        total_price *= 0.7 # 3割引き = 0.7倍
        print("3割引しました")
    elif int(num_apple) + int(num_grape) >= 5:
      total_price *= 0.8 # 2割引 = 0.8倍
      print("2割引しました")
    else:
      print("割引はありません")

    print("合計金額は" + str(int(total_price)) + "円です")

    running = input("プログラムを終了する場合は1以外の整数値を入力してください：") # ここで入力した値に応じて次の周回に入るかどうかを決める
    running = int(running) # inputだとデフォルトがstr型なのでint型に変換する必要がある

  except ValueError:
    print("半角整数値を入力してください")
    running = 1 # runningのinputで数値以外の値を入力した場合、1に戻さないと周回が終わってしまう

print("プログラムを終了します") # while文の外（インデントしない位置）にこれを書くことでwhile文から抜けたときに実行されるようになる
````

````python
# runningが1の間、繰り返されるwhile文を作る
fruits = ["apple", "grape"]
prices = {"apple":100, "grape":50}

running = 1 # whileの条件用の変数を定義（boolで定義してもok）
while running == 1:
  try:
    num_apple = input("appleの個数を入力してください：")
    num_grape = input("grapeの個数を入力してください：")

    if int(num_apple) + int(num_grape) >= 10:
      total_price = (prices["apple"] * int(num_apple) + prices["grape"] * int(num_grape)) * 0.7 # 3割引き = 0.7倍
      print("3割引しました")
    elif int(num_apple) + int(num_grape) >= 5:
      total_price = (prices["apple"] * int(num_apple) + prices["grape"] * int(num_grape)) * 0.8 # 2割引 = 0.8倍
      print("2割引しました")
    else:
      total_price = prices["apple"] * int(num_apple) + prices["grape"] * int(num_grape)
      print("割引はありません")

    print("合計金額は" + str(int(total_price)) + "円です")

    ############################## ここからが違う ##############################

    while True:
      running = input("プログラムを終了する場合は0、続ける場合は1を入力してください：") # ここで入力した値に応じて次の周回に入るかどうかを決める
      if running == '0':
        running = int(running) # inputだとデフォルトがstr型なのでint型に変換する必要がある
        print("それではプログラムを終了します")
        break
      elif running == '1':
        running = int(running) # inputだとデフォルトがstr型なのでint型に変換する必要がある
        print("それではプログラムを続けます")
        break
      else:
        print("0か1で回答してください")

    ############################################################################

  except ValueError:
    print("半角整数値を入力してください")
````
