# <font color = "blue">リスト型</font>
**list型**：データをまとめて扱うのに便利なデータ型  
リストを使うと、１つの変数名と「インデックス」という番号で値を管理している！   
**要素**：リスト内の個々の値のこと  

 - リスト名（変数名）の後ろに**[ ]**で囲って、まとめる  
 - リスト名[インデックスの番号]　→要素を指定して、取り出せる！  
　　<font color = "green">**注意**</font>：インデックスは「０」からスタートですべて整数値   




```python
# フルーツの種類が増えていくたびにいちいち書くのって面倒じゃない？？
fruit_1 = "apple"
fruit_1_price = 100 # 100円/個
fruit_1_quantity = 2.5 # 2.5個

fruit_2 = "grape"
fruit_2_price = 50 # 50円/個
fruit_2_quantity = 3 # 3個

# リスト（配列）にしてまとめることができる！
fruits = ["apple", "grape"] # 0番目が"apple"、1番目が"grape"になる（最初は0っていうのを意識しよう）

# リスト名[何番目]でリストの中身を取り出すことができる！
print("私は" + fruits[0] + "と" + fruits[1] + "を買います")
```

# <font color = "blue">辞書型</font>
**dictionary型**：キー（左側）と対応する値（右側）の<font color = "red">**ペア**</font>で管理するデータ型  
→listのインデックスの代わりにキー（文字列）によって要素を特定できるように!

 - 辞書の生成：｛キー１：値１, キー２：値２... }	<font color = "green">←かっこの形式注意！</font>  



```python
# 値段と買う量に関してもまとめたいなぁ
total_price = fruit_1_price * fruit_1_quantity + fruit_2_price * fruit_2_quantity

# そんなときは辞書が使える！
prices = {"apple":100, "grape":50} # apple番目のようなラベル付をすることでappleのときに呼び出せるようになる
quantities = {"apple":2.5, "grape":3}

total_price = prices["apple"] * quantities["apple"] + prices["grape"] * quantities["grape"]

print("合計金額は" + str(int(total_price)) + "円です")
```
