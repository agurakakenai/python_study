# <font color = "blue">グローバル変数</font>



前のセクションより：ローカル変数は書かれている関数内でしか使うことはできない  
でもどの関数でも共通して使いたい変数がある場合もある（参照用の変数とかでよく使います）  
そのときは<font color ="red">**グローバル変数** </font>を使う

### <font color = "blue">グローバル変数って？</font>
**グローバル変数**：プログラム全体で使うことができる変数のこと  
　　　　　　　　　　　任意の関数からアクセスできる  
<font color = "green">注意</font>:ローカル変数とグローバル変数に同じ名前の変数があると、関数の内部ではローカル関数が優先される！


````python
### 関数 ###
def total_caliculator (num_apple, num_grape):
  fruits = ["apple", "grape"]
  prices = {"apple":100, "grape":50}

  total_price = prices["apple"] * int(num_apple) + prices["grape"] * int(num_grape)
  if int(num_apple) + int(num_grape) >= 10:
      total_price *= 0.7 # 3割引き = 0.7倍
      print("3割引しました")
  elif int(num_apple) + int(num_grape) >= 5:
    total_price *= 0.8 # 2割引 = 0.8倍
    print("2割引しました")
  else:
    print("割引はありません")

  print("合計金額は" + str(int(total_price)) + "円です")


### メイン関数 ###
if __name__ == "__main__":
  running = 1 # whileの条件用の変数を定義（boolで定義してもok）
  while running == 1:
    try:
      num_apple = input("appleの個数を入力してください：")
      num_grape = input("grapeの個数を入力してください：")

      total_caliculator(num_apple, num_grape) # この1行で合計金額の計算ができる

      while True:
        running = input("プログラムを終了する場合は0、続ける場合は1を入力してください：") # ここで入力した値に応じて次の周回に入るかどうかを決める
        if running == '0':
          running = int(running) # inputだとデフォルトがstr型なのでint型に変換する必要がある
          print("それではプログラムを終了します")
          break
        elif running == '1':
          running = int(running) # inputだとデフォルトがstr型なのでint型に変換する必要がある
          print("それではプログラムを続けます")
          break
        else:
          print("0か1で回答してください")

    except ValueError:
      print("半角整数値を入力してください")
````

## <font color = "blue">グローバル変数の書き方</font>    

基本的なグローバル変数の書き方：
-  関数の外で変数を設定する  
- 変数を使いたいときは<font color = "red"> global 変数名 </font>と書く

---




```
value =1 #グローバル関数の設定  
def test():  
  global value    #globalで変数valueをグローバル変数として宣言
  value = 20      #グローバル変数のvalueに20を代入
  print(value)    #20が表示される

print(value)      #関数の外でも20と表示される
```





---




今回は

```
while True:
  ---
```
の部分を関数にしてrunningをグローバル変数にしてみましょう

````python
running = 1 # 関数の外に書くことでグローバル変数として定義される（メイン関数には入れてもグローバル変数として定義されるけど、今回はわかりやすいのでメイン関数からも外に書く）

### 関数 ###
def total_caliculator (num_apple, num_grape):
  fruits = ["apple", "grape"]
  prices = {"apple":100, "grape":50}

  total_price = prices["apple"] * int(num_apple) + prices["grape"] * int(num_grape)
  if int(num_apple) + int(num_grape) >= 10:
      total_price *= 0.7 # 3割引き = 0.7倍
      print("3割引しました")
  elif int(num_apple) + int(num_grape) >= 5:
    total_price *= 0.8 # 2割引 = 0.8倍
    print("2割引しました")
  else:
    print("割引はありません")

  print("合計金額は" + str(int(total_price)) + "円です")

### 関数 ###
def process_status():
  global running # この関数内で使うrunningはグローバル変数のrunningであるということを宣言する
  while True:
    running = input("プログラムを終了する場合は0、続ける場合は1を入力してください：") # ここで入力した値に応じて次の周回に入るかどうかを決める
    if running == '0':
      running = int(running) # ここでメイン関数の条件であるrunningに0が代入される
      print("それではプログラムを終了します")
      break
    elif running == '1':
      running = int(running)
      print("それではプログラムを続けます")
      break
    else:
      print("0か1で回答してください")

### メイン関数（メイン関数の中身はすべてグローバル変数になっている） ###
if __name__ == "__main__":
  while running == 1:
    try:
      num_apple = input("appleの個数を入力してください：")
      num_grape = input("grapeの個数を入力してください：")

      total_caliculator(num_apple, num_grape)
      process_status() # この関数の中でグローバル変数のrunningが更新されたら次のwhile文の条件がwhile==1ではなくなるので終了する

    except ValueError:
      print("半角整数値を入力してください")
````

process_status関数の中で<font color = "red">グローバル変数runningを更新できたから</font>メイン関数のwhile文のループから抜けられました！
