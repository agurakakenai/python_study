# <font color = "blue">戻り値</font>

**戻り値**：処理の結果として戻される値のこと  
　　　　関数での処理結果をメイン関数に戻したい（返したい）場合に使う  
グローバル変数は計算結果などのやり取りにはあまり使わない！（基本的にはグローバル変数は横着をしている）  

### <font color = "blue">戻り値の書き方</font>
戻したい変数を


```
return 戻したい変数名

```
としてコードを書いてください。


````python
running = 1

### 関数 ###
def total_caliculator (num_apple, num_grape):
  fruits = ["apple", "grape"]
  prices = {"apple":100, "grape":50}

  total_price = prices["apple"] * int(num_apple) + prices["grape"] * int(num_grape) # 今回はglobal変数ではないので関数内のtotal_priceはメイン関数のtotal_priceとは別もの
  if int(num_apple) + int(num_grape) >= 10:
      total_price *= 0.7 # 3割引き = 0.7倍
      print("3割引しました")
  elif int(num_apple) + int(num_grape) >= 5:
    total_price *= 0.8 # 2割引 = 0.8倍
    print("2割引しました")
  else:
    print("割引はありません")

  # print("合計金額は" + str(int(total_price)) + "円です") # ここのprint文を消して、メイン関数でprintしてください
  return total_price # total_priceの値を戻り値として渡す

### 関数 ###
def process_status():
  global running # この関数内で使うrunningはグローバル変数のrunningであるということを宣言する
  while True:
    running = input("プログラムを終了する場合は0、続ける場合は1を入力してください：") # ここで入力した値に応じて次の周回に入るかどうかを決める
    if running == '0':
      running = int(running) # ここでメイン関数の条件であるrunningに0が代入される
      print("それではプログラムを終了します")
      break
    elif running == '1':
      running = int(running)
      print("それではプログラムを続けます")
      break
    else:
      print("0か1で回答してください")

### メイン関数（メイン関数の中身はすべてグローバル変数になっている） ###
if __name__ == "__main__":
  while running == 1:
    try:
      num_apple = input("appleの個数を入力してください：")
      num_grape = input("grapeの個数を入力してください：")

      total_price = total_caliculator(num_apple, num_grape) # 関数の戻り値をメイン関数のtotal_priceに代入
      print("合計金額は" + str(int(total_price)) + "円です")

      process_status() # この関数の中でグローバル変数のrunningが更新されたら次のwhile文の条件がwhile==1ではなくなるので終了する

    except ValueError:
      print("半角整数値を入力してください")
````
