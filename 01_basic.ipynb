# <font color = "blue">型</font>
 データの種類ごとに<font color = "red">型</font>がある！

| 型           |    クラス    |         説明              |     例      |  
| :---:        |    :---:     |         :---:             |    :---:    |
| 文字列       |   **str**   | 文字列を管理するデータ型  | "hello", "a"|
| 整数型       |   **int**   | 整数を管理するデータ型    | 5, 100, -10 |  
| 浮動小数点型 |   **float** | 小数を管理するデータ型    | 3.14, -9.5  |

**type()関数**：()の中にある引数や値のクラスを表示する

---



````python
# 型はtype()をprintしてあげること確認できる！
fruit_1 = "apple"
fruit_1_price = 100 # 100円/個
fruit_1_quantity = 2.5 # 2.5個

fruit_2 = "grape"
fruit_2_price = 50 # 50円/個
fruit_2_quantity = 3 # 3個

print(type(fruit_1)) # 文字列（string）のstr型
print(type(fruit_1_price)) # 整数（integer）のint型
print(type(fruit_1_quantity)) # 小数（float）のfloat型
````

### <font color = "blue">型の相互関係</font>
型が一緒なら、変数を用いて計算できる！  
型がごちゃまぜになっているとエラーが起こってしまう   
<font color = "red">型に気を付けてコードを書こう！</font>



````python
# 型はtype()をprintしてあげること確認できる！
fruit_1 = "apple"
fruit_1_price = 100 # 100円/個
fruit_1_quantity = 2.5 # 2.5個

fruit_2 = "grape"
fruit_2_price = 50 # 50円/個
fruit_2_quantity = 3 # 3個

# 型が文字列同士の場合は連結ができる！
print("私は" + fruit_1 + "と" + fruit_2 + "を買います")

# 型が数値の場合計算ができる！
total_price = fruit_1_price * fruit_1_quantity + fruit_2_price * fruit_2_quantity # appleは100円/個 * 2.5個、grapeは50円/個 * 3個だから合計は350円
print(total_price)

# せっかくならちゃんと説明付きでprintしたい！
print("合計金額は" + total_price + "円です") # でもこれだとここでエラーになる！（文字列と数値がごちゃまぜになっているから！）
````

### <font color = "blue">型の相互変換</font>
「コンストラクタ」と呼ばれるオブジェクトを生成するための関数を使う！
float()：浮動小数点型に変換  
int()：整数型に変換  
str()：文字列に変換  

````python
# それなら文字列に変換してやればいいじゃないか！
print("合計金額は" + str(total_price) + "円です") # strで囲ってやることで数値を文字列に変換できる！

# でもこのままだと400.0円って小数が出てきてキモい、、、
# これも同じ要領でintで囲ってやればいい！
total_price = int(total_price) # total_priceをint型（整数）に変換
print("合計金額は" + str(total_price) + "円です")  # 直前に整数に変換したから400円になる！

# ちなみにもちろん逆もできるよ！（キモいけど）
total_price = float(total_price) # total_priceをfloat型（小数）に変換
print("合計金額は" + str(total_price) + "円です")  # 直前に整数に変換したから400円になる！
````


# <font color = "blue">四則演算</font>




##【基本的な算術演算子】
- **+**：足し算  
- **-**：引き算  
- **＊**：かけ算  
- **/**：割り算  
- **//**：商（割り算の結果として整数部分を求める）  
- **%**：余り（割り算の割り切れない数）  
- **＊＊**: べき乗（a＊＊bでbでaのb乗を表す）

 <font color = "green">**注意**</font> :
割り算はfloat型で返される
商は計算式にfloat型を含むときに答えが整数でもfloat型
で返す  
　　　べき乗では　0の0乗は1　になる！



##【代入演算子】
- **+=** : a = a + bと同じ
- **-=** : a = a -bと同じ
- ***=** : a = a * bと同じ
- **/=** : a = a / ｂと同じ  



##【演算の優先順位】
- かけ算・割り算は足し算・引き算よりも優先される  
- 優先順位の変更には（）を使う
<u>※数学と一緒</u>

````python
# 基本的な演算
a = 5
b = 2

# 足し算 +
x = a + b
print(str(a) + " + " + str(b) + " = " + str(x))

# 引き算 -
x = a - b
print(str(a) + " - " + str(b) + " = " + str(x))

# 掛け算 *
x = a * b
print(str(a) + " * " + str(b) + " = " + str(x))

# 割り算 /
x = a / b
print(str(a) + " / " + str(b) + " = " + str(x))

# 商 //
x = a // b
print(str(a) + " // " + str(b) + " = " + str(x))

# 余剰（あまり） %
x = a % b
print(str(a) + " % " + str(b) + " = " + str(x))

# べき乗（aのb乗）**
x = a ** b
print(str(a) + " ** " + str(b) + " = " + str(x))
````

# <font color ="blue">ライブラリを使った演算</font>

Pythonのファイル = 「〜.py」ファイルのこと  
**モジュール**:他のPythonプログラムからimportして使うことができる「〜.py」のファイル   
↓  
パッケージ：複数のモジュールをまとめたもの   
↓   
**ライブラリ**：複数のパッケージをまとめてインストールできるようにしたもの（関数、モジュール、パッケージを総称して呼ぶこともあるそう）  
→<u>これまでに出てきたintなどの型、print()などの関数などをまとめたモジュールというものがセットに！</u>

**標準ライブラリ**：デフォルトで含まれるライブラリ（標準装備されているもの）






### <font color ="blue">コードの書き方</font>
モジュールのインポート：

```
import モジュール名
```
モジュール内の指定した関数を使う：

```
from モジュール名 import 関数名１, 関数名２, …
```

````python
# 半径(radius)が4の円の面積を求めたい！
# 円の面積は円周率 * 半径の2乗だから
r = 4 # radius
S = 3.14 * r ** 2
print("面積は" + str(S) + "です") # でも円周率3.14って正確じゃないよね（円周率は3.141592...）
````

````python
# もっと正確に計算したい！
# でもそのままだとより正確な円周率は使えない、、、
# なら必要なライブラリを追加してその中の円周率の値を使おう！

import numpy as np # numpyという数学のライブラリをnpという名前で呼び出せるように追加する

r = 4 # radius
S = np.pi * r ** 2 # np.piってやることでnp（numpy）には円周率piを呼び出せるよ！（ライブラリ名.ライブラリ内の使いたいもの）
print("面積は" + str(S) + "です") # より正確な値が出せた！
````
