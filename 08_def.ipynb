# <font color = "blue">関数</font>

**関数**：処理手順（アルゴリズム）を切り分けることが可能！



- 切り分けることで得られるメリット

  - 共通の処理手順（アルゴリズム）があるときにいちいち書かなくていい  
  - 手直しするときに関数を直すだけで大丈夫  
（実際共通部分が1つ1つ分けられていると、自分が直している場所が直したい場所と違って作業時間が余計にかかってしまうなんてこともある）

  そんなときには  
  **関数**を作って、<font color ="red">アルゴリズムを抜き出して、それぞれのパラメータ（引数）によって処理が変わるように</font>すればいい

```python
fruits = ["apple", "grape"]
prices = {"apple":100, "grape":50}

running = 1
while running == 1:
  try:
    num_apple = input("appleの個数を入力してください：")
    num_grape = input("grapeの個数を入力してください：")

    ### ここのtotal_priceの計算部分を関数として切り分ける ###
    total_price = prices["apple"] * int(num_apple) + prices["grape"] * int(num_grape)
    if int(num_apple) + int(num_grape) >= 10:
        total_price *= 0.7 # 3割引き = 0.7倍
        print("3割引しました")
    elif int(num_apple) + int(num_grape) >= 5:
      total_price *= 0.8 # 2割引 = 0.8倍
      print("2割引しました")
    else:
      print("割引はありません")

    print("合計金額は" + str(int(total_price)) + "円です")
    ##########################################################

    while True:
      running = input("プログラムを終了する場合は0、続ける場合は1を入力してください：")
      if running == '0':
        running = int(running)
        print("それではプログラムを終了します")
        break
      elif running == '1':
        running = int(running)
        print("それではプログラムを続けます")
        break
      else:
        print("0か1で回答してください")

  except ValueError:
    print("半角整数値を入力してください")
```

## <font color = "blue">関数の書き方</font>

### <font color = "blue">def文</font>

**def文**:関数を定義することができる  
　　　関数本体はインデントを使ってブロックにする  
　　　関数が値を返す時は　**return　値（戻り値）**　で戻り値を指定する

<font color ="green">注意</font>:関数名の付け方は変数名と同じ  
 - 基本的に、英小文字
 - 複数の単語をつなげるときは＿(アンダースコア)

### <font color = "blue">関数の取り扱い</font>

**ローカル変数**：  
関数内部で値を代入した変数は、その関数の内部でのみ使うことができる  
関数の外部からアクセスできない  
<font color = "red">変数の有効な範囲を意識しないといけない</font>  


<div>

</div>

```python
### 関数 ###
def total_caliculator (num_apple, num_grape):
  fruits = ["apple", "grape"]
  prices = {"apple":100, "grape":50}

  total_price = prices["apple"] * int(num_apple) + prices["grape"] * int(num_grape)
  if int(num_apple) + int(num_grape) >= 10:
      total_price *= 0.7 # 3割引き = 0.7倍
      print("3割引しました")
  elif int(num_apple) + int(num_grape) >= 5:
    total_price *= 0.8 # 2割引 = 0.8倍
    print("2割引しました")
  else:
    print("割引はありません")

  print("合計金額は" + str(int(total_price)) + "円です")

### メイン関数(これはifを使っているけど例外中の例外なので、メイン関数はこういう書き方をするんだと覚えてください) ###
if __name__ == "__main__":
  fruits = ["apple", "grape"]
  prices = {"apple":100, "grape":50}

  running = 1 # whileの条件用の変数を定義（boolで定義してもok）
  while running == 1:
    try:

      num_apple = input("appleの個数を入力してください：")
      num_grape = input("grapeの個数を入力してください：")

      total_caliculator(num_apple, num_grape) # この1行で合計金額の計算ができてしまう

      while True:
        running = input("プログラムを終了する場合は0、続ける場合は1を入力してください：") # ここで入力した値に応じて次の周回に入るかどうかを決める
        if running == '0':
          running = int(running) # inputだとデフォルトがstr型なのでint型に変換する必要がある
          print("それではプログラムを終了します")
          break
        elif running == '1':
          running = int(running) # inputだとデフォルトがstr型なのでint型に変換する必要がある
          print("それではプログラムを続けます")
          break
        else:
          print("0か1で回答してください")

    except ValueError:
      print("半角整数値を入力してください")
```
