# <font color = "blue">例外処理：try, except</font>







下のコードはtry, exceptを使っていないため、  
半角整数値以外が入力されると、エラーを表示する！

<font color = "red">下のプログラムを実行し、半角整数値以外を入力し,エラーが起こるか確かめてみよう。</font>

````python
# 前のセッションでやった個数の部分を実行中に入力できるようにする
fruits = ["apple", "grape"]
prices = {"apple":100, "grape":50}

num_apple = input("appleの個数を入力してください：")
num_grape = input("grapeの個数を入力してください：")

total_price = prices["apple"] * int(num_apple) + prices["grape"] * int(num_grape)
if int(num_apple) + int(num_grape) >= 10:
    total_price *= 0.7 # 3割引き = 0.7倍
    print("3割引しました")
elif int(num_apple) + int(num_grape) >= 5:
  total_price *= 0.8 # 2割引 = 0.8倍
  print("2割引しました")
else:
  print("割引はありません")

print("合計金額は" + str(int(total_price)) + "円です")

````

### <font color = "blue">try, exceptを使うとき</font>
事前にユーザがどんなエラーが出しうるか分かっていて、そのエラーが出たときに例外処理をしたいときに使う！

今回のパターンだとユーザがinput関数にintに変換できるもの以外を入れてしまうとnum_appleとnum_grapeで<u>ValueError</u>が出てしまう！  
でもこのValueErrorが出ることは想定内

だから…  
ValueErrorが出た時に例外的に対応をできるようにしたい！　  
⇒ <font color ="red">try, exceptを使って、例外処理をする</font>  


### <font color = "blue">try, exceptの書き方</font>
if文の書き方と同様に考えていけばOK
```
try:
  例外が発生するかもしれないけど、実行したい処理
except エラー名:
  例外に対して行う処理
```
<font color = "green">**注意**</font> : インデントが正しくできているかチェック！

````python
fruits = ["apple", "grape"]
prices = {"apple":100, "grape":50}

# とりあえずtryの中身を実行して、エラーが出たらexceptに移動する
# exceptで指定したエラーとtry中に出たエラーが同じ場合、except内部のプログラムを実行する
try:
  num_apple = input("appleの個数を入力してください：")
  num_grape = input("grapeの個数を入力してください：")

  total_price = prices["apple"] * int(num_apple) + prices["grape"] * int(num_grape)
  if int(num_apple) + int(num_grape) >= 10:
      total_price *= 0.7 # 3割引き = 0.7倍
      print("3割引しました")
  elif int(num_apple) + int(num_grape) >= 5:
    total_price *= 0.8 # 2割引 = 0.8倍
    print("2割引しました")
  else:
    print("割引はありません")

  print("合計金額は" + str(int(total_price)) + "円です")

except ValueError:
  print("半角整数値を入力してください")
````
