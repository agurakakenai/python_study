#<font color = "blue">for文</font>
**for文** : 回数が決まっているときに便利な繰り返し処理のこと
  
イテレート可能なオブジェクト：オブジェクトから要素を<font color = "red">一つずつ</font>取り出して処理を繰り返しおこなうことができるもの  
ex)リスト、辞書、文字列、、、  




````python
fruits = ["apple", "grape"]
prices = {"apple":100, "grape":50}
quantities = {"apple":2.5, "grape":3}

fruits.append("orange")
prices.update({"orange":200, "peach":150})
quantities.update({"orange":1})

print(fruits)
print(prices)
print(quantities)

````

````python
# 要素ができるたびにこの式が長くなっていくのは大変じゃない？
total_price = prices["apple"] * quantities["apple"] + prices["grape"] * quantities["grape"] + prices["orange"] * quantities["orange"]
print("合計金額は" + str(total_price) + "円です")
````

## <font color = "blue">for文の書き方</font>
書き方：
```
# for 変数 in　イテレート可能なオブジェクト：
	処理１
	処理２
	…
```

<font color = "green">**注意**</font>：  
要素の順序の通りに処理をしていく   
リストや辞書は要素を順番に取り出す！  
文字列は最初から１文字ずつ取り出す！  
<u>インデント（字下げ）に気を付けよう</u>  


````python
# ならfor文をつかって繰り返しちゃえばいい！
total_price = 0 # 合計値も最初は0

for fruit in fruits:
  # for文（繰り返したい部分）は字下げする！
  total_price += prices[fruit] * quantities[fruit] # fruitにはfruitsリストの左から順番に代入される！

print("合計金額は" + str(total_price) + "円です")
````
