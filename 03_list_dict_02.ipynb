### <font color = "blue">append関数：リストに要素を追加</font>
**append()**：<font color = "red">末尾</font> に要素を追加する  
　　　　　一度に<font color = "red">１つ</font>の要素 を追加



### <font color = "blue">update関数：辞書に要素を追加</font>
**update()**：<font color = "red">末尾</font>に別の辞書を追加する   
　　　　　一度に<font color = "red">複数</font>の要素を追加できる



```python
# フルーツにorangeを追加したいときってあるよね
fruits = ["apple", "grape"]
prices = {"apple":100, "grape":50}
quantities = {"apple":2.5, "grape":3}

print(fruits)
print(prices)
print(quantities)
print('\n') #手動で改行するやつ

fruits.append("orange") # リストの場合はappend（追加したい要素）、1度に1つの要素しか追加できない
# fruits += ["orange"]でもいける（この場合はリスト同士を連結することを意味するので複数の要素を追加できる）

prices.update({"orange":200, "peach":150}) # 辞書の場合はupdate（追加したい辞書）、1度に複数の要素を追加できる
quantities.update({"orange":1}) # もちろん1つの要素でもいい（ただし辞書型で書く必要はある）
# prices += {"orange":200}ではいけない

print(fruits)
print(prices)
print(quantities)
```

```python
# orangeも計算に加えてみよう
total_price = prices["apple"] * quantities["apple"] + prices["grape"] * quantities["grape"] + prices["orange"] * quantities["orange"]

print("合計金額は" + str(int(total_price)) + "円です")


```
